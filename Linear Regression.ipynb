```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
%matplotlib inline
plt.scatter(df.area , df.price, color="red", marker="*")
plt.xlabel("Area (Sq. Ft)", fontsize=12,fontweight="bold")
plt.ylabel("Price (USD)",fontsize=12,fontweight="bold")
plt.title("Linear Regression",fontsize=17,fontweight="bold")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[4000]])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Perform linear regression
coefficients = np.polyfit(x, y, 1)
m = coefficients[0]  # slope
c = coefficients[1]  # y-intercept

# Create scatter plot
plt.scatter(x, y)

# Generate line coordinates
x_line = np.array([min(x), max(x)])
y_line = m * x_line + c

# Plot the linear regression line
plt.plot(x_line, y_line, color='red', label='Linear Regression Line')

# Add the equation of the regression line
equation = f'y = {m:.2f}x + {c:.2f}'  # Format the equation string
plt.annotate(equation, xy=(0.6, 0.95), xycoords='axes fraction', fontsize=12)

# Set labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Display the plot
plt.show()
```
