```python
import pandas as pd 
import numpy as np 
import seaborn as sns
from sklearn import linear_model
```

```python
#df = pd.read_excel("data_Reg.xlsx",sheet_name="Sheet2")
df = pd.read_excel("dataIshan.xlsx",sheet_name="Sheet2")
df
```

```python
# Handling missing values (NaN or NA)
md_bedroom = df.bedroom.median()
md_bedroom
```

```python
df.bedroom =df.bedroom.fillna(md_bedroom)
df
```

```python
# Visualizing the relationships between features using pair plots 
sns_plot = sns.pairplot(df,hue='Species',height=2,diag_kind='kde',kind='reg',plot_kws={'line_kws':{'color':'red'}})
sns_plot.savefig('pairplot.png',dpi=300)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedroom','age']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]])
```
